```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))

plt.hist(
    df["Age"],
    bins=10,
    color="PURPLE",
    edgecolor="black"
)

plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```
